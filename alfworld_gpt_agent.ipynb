```python
## Expert Model Class

class ExpertActionPredictor():
    def __init__(self,agent, predict_func):
        self.agent=agent
        self.predict_func=predict_func
    
    def predict(self, obs, info):
        return self.predict_func(obs, info, agent=self.agent)
```

```python
## Extract from ALFWorld DAgger IL

def alfworld_action_predictor(ob, info, agent):
    first_sight_string = agent.first_sight_string
    task_desc_string = agent.task_desc_string
    dynamic = agent.dynamic
    execute_action = agent.execute_action

    observation_string = list(ob)
    ## We can anyway to process the obs like this 
    _, observation_string = agent.get_task_and_obs(observation_string)
    observation_string = agent.preprocess_observation(observation_string)

    if agent.action_space == "exhaustive":
        action_candidate_list = [extract_admissible_commands(intro, ob) for intro, ob in zip(first_sight_string, observation_string)]
    else:
        action_candidate_list = list(info["admissible_commands"])
    action_candidate_list = agent.preprocess_action_candidates(action_candidate_list)
    observation_string = [item + " [SEP] " + a for item, a in zip(observation_string, execute_action)]  # appending the chosen action at previous step into the observation
    print(observation_string)

    # push obs into observation pool
    agent.observation_pool.push_batch(observation_string)
    # get most recent k observations
    most_recent_observation_string = agent.observation_pool.get()
    print(most_recent_observation_string)
    # predict actions
    if agent.action_space == "generation":
        # heuristically unstick the agent from generating the same thing over and over again
        prev_actions = copy.copy(execute_action)
        if agent.unstick_by_beam_search:
            if "Nothing happens" in observation_string[0]:
                agent.smart[0]["not working"].append(execute_action[0])

        execute_action, current_dynamic = agent.command_generation_greedy_generation(most_recent_observation_string, task_desc_string, dynamic)

        # heuristically unstick the agent from generating the same thing over and over again
        if agent.unstick_by_beam_search:
            if "Nothing happens" in observation_string[0] and execute_action[0] in agent.smart[0]["not working"]:
                if len(agent.smart[0]["to try"]) == 0:
                    bs_action, _ = agent.command_generation_beam_search_generation(most_recent_observation_string[0: 1], task_desc_string[0: 1], None if dynamic is None else dynamic[0: 1])
                    bs_action = bs_action[0]
                    agent.smart[0]["to try"] += bs_action

                agent.smart[0]["to try"] = [item for item in agent.smart[0]["to try"] if item != prev_actions[0]]
                if len(agent.smart[0]["to try"]) > 0:
                    execute_action[0] = agent.smart[i]["to try"][0]
            else:
                agent.smart[0] = {"not working": [], "to try": []}  # reset

    elif agent.action_space in ["admissible", "exhaustive"]:
        execute_action, _, current_dynamic = agent.admissible_commands_greedy_generation(most_recent_observation_string, task_desc_string, action_candidate_list, dynamic)
    else:
        raise NotImplementedError()

    agent.dynamic = current_dynamic
    agent.execute_action = execute_action

    return execute_action
```

```python
from langchain.utilities import SerpAPIWrapper
from langchain.agents import Tool
from langchain.tools.file_management.write import WriteFileTool
from langchain.tools.file_management.read import ReadFileTool

from output_parser import (
    AutoGPTOutputParser,
    BaseAutoGPTOutputParser,
)
from prompt import AutoGPTPrompt
from langchain.chains.llm import LLMChain


DELIMITER="%%%%"     #Pick an unlikely delimiter used somewhere else

def extract_admissible_commands(info: dict) -> list:

    return info["admissible_commands"][0]


def process_ob(ob):
    if ob.startswith('You arrive at loc '):
        ob = ob[ob.find('. ')+2:]    
    return ob


def alfworld_action(action: str) -> str:
    global env, init_spec, task_spec
    observation, reward, done, info = env.step([action])
    observation, reward, done = process_ob(observation[0]), info['won'][0], done[0]
    obs_spec = f'After you {action}, current observation is: {observation}\n'
    obs_prompt = init_spec + task_spec + obs_spec

    reward = int(reward)
    if done:
        return f"{obs_prompt} {{'reward':{reward}}}"      

    info['alfworld_ob'] = observation
    admissible_commands_spec = "Available Actions are:\n" + '\n'.join(['[action] '+ ac + ' [action_]' for ac in info["admissible_commands"][0]]) 
    return obs_prompt + admissible_commands_spec + DELIMITER + json.dumps(info)


def init_vectorstore(embedding_size=1536):
    index = faiss.IndexFlatL2(embedding_size)
    vectorstore = FAISS(embeddings_model.embed_query, index, InMemoryDocstore({}), {})
    return vectorstore
```

```python
import autogpt
from importlib import reload
reload(autogpt)
from autogpt import AutoGPT

LOOP_LIMIT = 35
input_variables=["memory", "messages", "goals", "user_input"]
def init_agent(ai_name, ai_role, tools, input_variables,
               llm, vectorstore, init_obs = None, verbose = True, expert_predictor=None):
    prompt = AutoGPTPrompt(
                ai_name=ai_name,
                ai_role=ai_role,
                tools=tools,
                input_variables=input_variables,
                token_counter=llm.get_num_tokens
            )
    print(prompt)
    chain = LLMChain(llm=llm, prompt=prompt)


    agent = AutoGPT(
        ai_name="TextWorld Expert",
        memory=vectorstore.as_retriever(),
        chain=chain,
        output_parser=AutoGPTOutputParser(),
        tools=tools,
        loop_limit=LOOP_LIMIT,
        init_obs=init_obs,
        expert_predictor=expert_predictor,
    )
    agent.chain.verbose = verbose
    return agent
```

```python
def write_dict_to_file(dictionary, file_path):
    # Open the file in write mode
    with open(file_path, "w") as file:
        for line in dictionary:
            file.write(str(line) + "\n")
```
